```python
tpf = TessTargetPixelFile('https://archive.stsci.edu/hlsps/tess-data-alerts/hlsp_tess-data-alerts_tess_phot_00263003176-s01_tess_v1_tp.fits')
```

```python
median_image = np.nanmedian(tpf.flux, axis=0)
```

```python
aperture_mask = median_image> np.nanpercentile(median_image,85)
```

```python
tpf.plot(aperture_mask=aperture_mask)
```

```python
lc =tpf.to_lightcurve(aperture_mask=aperture_mask)
lc.errorbar()
```

```python
import lightkurve as lk
%matplotlib inline
```

```python
search_result = lk.search_lightcurve('Kepler-69', author='Kepler', cadence='long')
lc_collection = search_result.download_all()
lc_collection.plot()
```

```python
search_result
```

```python
lc =lc_collection.stitch().flatten(window_length=901).remove_outliers()
lc.plot()
```

```python
import numpy as np
period = np.linspace(1,20,10000)
bls = lc.to_periodogram(method='bls',period=period, frequency_factor =500) 
bls.plot()
```

```python
planet_b_period = bls.period_at_max_power
planet_b_t0 = bls.transit_time_at_max_power
planet_b_dur = bls.duration_at_max_power

planet_b_period
```

```python
ax =lc.fold(period=planet_b_period, epoch_time=planet_b_t0).scatter()
ax.set_xlim(-5,5)
```

```python
planet_b_mask = bls.get_transit_mask(period=planet_b_period, transit_time=planet_b_t0, duration = planet_b_dur)
```

```python
masked_lc = lc[~planet_b_mask]
ax = masked_lc.scatter()
lc[planet_b_mask].scatter(ax=ax,c='r', label='Masked')
```

```python
planet_b_model = bls.get_transit_model(period = planet_b_period, transit_time=planet_b_t0,duration = planet_b_dur)
```

```python
ax = lc.fold(planet_b_period,planet_b_t0).scatter()
planet_b_model.fold(planet_b_period, planet_b_t0).plot(ax=ax, c='r', lw=2)
ax.set_xlim(-5, 5)
```

```python
period = np.linspace(1, 300, 10000)
bls = masked_lc.to_periodogram('bls', period=period, frequency_factor=500)
bls.plot();
```

```python
planet_c_period = bls.period_at_max_power
planet_c_t0 = bls.transit_time_at_max_power
planet_c_dur = bls.duration_at_max_power

planet_c_period
```

```python
ax = masked_lc.fold(planet_c_period, planet_c_t0).scatter()
masked_lc.fold(planet_c_period, planet_c_t0).bin(.1).plot(ax=ax, c='r', lw=2,
                                                          label='Binned Flux')
ax.set_xlim(-5, 5);
```

```python
planet_c_model = bls.get_transit_model(period=planet_c_period,
                                       transit_time=planet_c_t0,
                                       duration=planet_c_dur)
```

```python
ax = lc.scatter();
planet_b_model.plot(ax=ax, c='dodgerblue', label='Planet b Transit Model');
planet_c_model.plot(ax=ax, c='r', label='Planet c Transit Model');
```

```python
from lightkurve import TessTargetPixelFile
import numpy as np

%matplotlib inline

tpf = TessTargetPixelFile('https://archive.stsci.edu/hlsps/tess-data-alerts/hlsp_tess-data-alerts_tess_phot_00012862099-s03_tess_v1_tp.fits')

tpf.plot(aperture_mask=tpf.pipeline_mask)

median_image =np.nanmedian(tpf.flux, axis=0)
```

```python
aperture_mask = median_image> np.nanpercentile(median_image,85)
```

```python
tpf.plot(aperture_mask=aperture_mask)
```

```python
lc=tpf.to_lightcurve(aperture_mask=aperture_mask)
lc.plot()
lc.errorbar()
```

```python
print(lc.time.shape)
```

```python
flat,trend =lc.flatten(window_length=1001, return_trend=True)
ax = lc.errorbar()
trend.plot(ax=ax, color='red',lw=2, label = 'Trend')
```
